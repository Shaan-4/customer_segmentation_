```python

clustering_data = df[['Age', 'Income', 'Total_Spending', 'Children']]


clustering_data['Income'] = clustering_data['Income'].fillna(clustering_data['Income'].median())

print("Data selected for model. Shape:", clustering_data.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
# This creates a numpy array of scaled numbers
scaled_array = scaler.fit_transform(clustering_data)

# Turn it back into a DataFrame so we can see it clearly
df_scaled = pd.DataFrame(scaled_array, columns=clustering_data.columns)
df_scaled.head()
```

```python
from sklearn.cluster import KMeans

wcss = [] # Within-Cluster Sum of Squares (Total error)
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method to find Optimal Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Replace '4' with the number you saw in the elbow plot
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42)
df['Cluster'] = kmeans.fit_predict(df_scaled)

# Let's see our segments!
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Income', y='Total_Spending', hue='Cluster', palette='viridis')
plt.title('Final Customer Segments')
plt.show()
```
